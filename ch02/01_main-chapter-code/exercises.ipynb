```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
original_str = "Akwirw ier"
```

```python
integers = tokenizer.encode(original_str)

print(integers)
```

```python
decoded_str = tokenizer.decode(integers)

print(decoded_str)
```
